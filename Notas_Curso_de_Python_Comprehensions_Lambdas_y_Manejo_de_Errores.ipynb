# ¿Qué son las list comprehensions?

Básicamente son una forma de crear listas de una manera elegante simplificando el código al máximo. 

````python
numeros = [1, 2, 34, 86, 4, 5, 99, 890, 45]
pares = []
for num in numeros:
    if num % 2 == 0:
        pares.append(num)
print(pares)
````

El código anterior crea una lista de números pares a partir de una lista de números. Para este tipo de situaciones es ideal el uso de las list comprehensions. 

El código anterior se puede modificar de la siguiente manera con la sintaxis de las list comprehensions:

````python
pares = [num for num in numeros if num % 2 == 0]
print(pares)
````

La estructura de las list comprehensions es la siguiente:

    [element for element in iterable if condition]

* **element**: Representa a cada uno de los elementos a poner en la nueva lista
* **for element in iterable**: Ciclo a partir del cual se extraerán elementos de otra lista o cualquier iterable
* **if condition**: Condición *opcional* para filtrar los elementos del ciclo

# ¿Qué son los dictionary comprehensions?

Se comporta igual que las list comprehensions, pero con la diferencia de que se guardan llaves y valores.



La estructura de los dict compregensions es la siguiente:
   
    {key:value for value in iterable if condition}

* **key:value**: Representa a cada una de las llaves y valores a poner en el nuevo diccionario
* **for value in iterable**: Ciclo a partir del cual se extraerán elementos cualquier iterable
* **if condition**: Condición opcional para filtrar los elementos del ciclo

Para cada elemento en un iterable, se coloca una llave y un valor solo si se cumple la condición. 

Por ejemplo:

    {i: i**3 for i in range(1, 101) if i % 3 != 0}
    
Para cada i en el rango de 1 hasta 100 se va a guardar i como la llave y a i al cubo como su valor, si y solo si, i es divisible por 3.    
```



````python
div_3 = {i: i**3 for i in range(1, 101) if i % 3 == 0}
for i in div_3:
    print(i)
````

# Funciones de Orden Superior (High Order functions)

Es una función que recibe como parámetro a otra función. Tenemos como ejemplo el siguiente código:


````python
def saludo(func):
    func()
    
def hola():
    print("Hola!!!")

def adios():
    print("Adios!!")
    
saludo(hola)
saludo(adios)
````

La función saludo toma como parámetro a las funciones hola y adios, dando como resultado que se impriman sus respectivos mensajes. 

Existen tres funciones de orden superior: filter, map y reduce.


# Función filter
Supongamos que tienes la primera lista pero sólo quieres los números impares 

    [1, 4, 5, 6, 9, 13, 19, 21]
	[1, 5, 9, 13, 19, 21]

Puedes hacerlo de la manera típica con una list comprehension de la siguiente manera 


````python
my_list = [1, 4, 5, 6, 9, 13, 19, 21]
add = [i for i in my_list if i %2 != 0]
print(add)
````

Esto mismo se puede resolver utilizando filter


````python
my_list = [1, 4, 5, 6, 9, 13, 19, 21]
add = list(filter(lambda x: x%2 != 0, my_list))
print(add)
````

Filter recibe un primer parámetro que es una función filtro. En este caso es lambda que recibe como parámetro x y retorna x%2 != 0. El segundo parámetro es un iterable (lista, dupla, etc), en este caso my_list. La función filter retorna un iterador. 

Su sintaxis es:

    filter(<funcion filtro>, <iterable>)
    
Volviendo al ejemplo, la función filter se coloca dentro de la función list para convertir todo en una lista. 

# Función map

Supongamos que tienes la siguiente lista pero quieres elevarlos al cuadrado:

    [1, 2, 3, 4 ,5]
    [1, 4, 9, 16, 25]

Haciéndolo con una list comprehensión sería de la siguiente manera


````python
my_list = [1, 2, 3, 4 ,5]
squares = list(map(lambda x:x**x,my_list))
print(squares)
````

Al igual que filter recibe una función anónima y un iterable como parámetros pero en este caso map ejecuta la función sobre cada uno de los elementos del iterable, sintaxis: 

    map(<funcion>, <iterable>)

# Función reduce 

Supongamos que tenemos el siguiente código el cual multiplica todos los elementos de la lista dando como resultado 32.


````python
my_list = [2, 2, 2, 2, 2]
all_mult = 1
for i in my_list:
    all_mult = all_mult * i
print(all_mult)    
````

Lo anterior se puede hacer con la función reduce. Tenemos que importar esta función desde functools para poder usarla, tiene los mismos argumentos que las anteriores funciones, reduce el iterable por medio de la función anónima, su sintaxis es: 

    reduce(<funcion reduccion>, <iterable>)
    
La función de reducción necesita de dos parámetros, uno que almacena el resultado (o el primer valor del iterable) y otro que opera con el siguiente valor del iterable:

    lambda a,b: <expresión>
    
Veamos el siguiente ejemplo:


````python
from functools import reduce
my_list = [2, 2, 2, 2, 2]
all_mult = reduce(lambda a,b: a * b, my_list)
print(all_mult)
````

La variable a  y  b irán cambiando conforme cada iteración se vaya realizando. La primera iteración a = 2 y b = 2 por lo que el resultado sería 4 y este pasaría a ser el primer elemento de la lista. La segunda iteración a = 4 y b = 2, pasando a ser 8 el primer elemento para la siguiente iteración y así sucesivamente hasta terminar con todos.

# Los errores en el código

Existen dos tios de errores en Python:

* **SyntaxError**: En inglés se conoce como un "typo", esto es cuando escribes mal algún comando. Aquí Python ni ejecuta el código porque antes de correrlo revisa si existen este tipo de errores en el programa antes de ejecutarlo.
* **Exception**: A diferencia de los SyntaxError, aquí Python si corrió el programa, pero algún fallo en la lógica hizo que el programa se detuviera en ese punto en específico. Existen diferentes tipos dentro de esta categoría algunas son:
    * KeyboardInterrupt: Sucede cuando precionas Ctrl + C
    * KeyError: Cuando intentas acceder a una llave que no existe dentro de un diccionario
    * IndexError: Cuando intentas acceder a un índice que no existe dentro de una lista
    * FileNotFoundError:  Cuando intentas abrir un archivo que no existe
    * ZeroDivisionError: Cuando intentas dividir entre cero
    * ImportError: Cuando intentas importar un módulo y existe algún error dentro de él.
    




Python muestra las excepciones a través de un mensaje llamado **Traceback**. Lo correcto, como programador, es leer este mensaje desde el final hasta el inicio. 

La sintaxis de un mensaje **traceback** sería como el siguiente:   

    Traceback (most recent call last):
        File "<stdin>", line 1, in <module> 
    ZeroDivisionError: division by zero

Este tiene los siguientes elementos:

* **ZeroDivisionError**: En este caso nos indica que es un error de división entre cero.
* **module**: Te indica en qué módulo existe el error
* **line #**: Te indica específicamente en qué línea del código existe el error
* **File**: Te indica dentro de qué archivo existe el error, en este caso stdin se refiere a la consola.
* **Traceback (most recent call last)**:significa que la llamada más reciente es la última (el programa se cerró después de esa llamada, se genero un error)

Cuando tenemos una excepción en python lo que sucede es que se crea un objeto de tipo exception que se va moviendo a través de los bloques de código hasta llegar al bloque principal si es que no se maneja dicha excepción en algún bloque intermedio el programa se interrumpe y genera el traceback

# Manejo de Excepciones

Supongamos que tenemos el siguiente código

    def palindrome(string):
        return string == string[::-1]

    print(palindrome(1))   
    
Al ejectuarlo nos dará el siguiente error:

        return string == string[::-1]

Entonces analicemos el mensaje. Nos dice que hay un **TypeError**, que se encuentra en el archivo **main.pu** en la **línea 2** la función **palindrome**, y también en la **línea 4** en **module**


Podemos anticipar este tipo de errores y hacer que Python corra algún código en caso de que se presente. Para ello utilizamos **tryexcept**.

* **try except**: Anidamos nuestro programa en dos bloques de código, el primero es el programa per se (el que se ejecuta normalmente, sin errores) y el segundo representa las instrucciones a seguir en caso de error. 

Su sintaxis es: 

    try:
        <bloque1>
    except <error> as <alias>:
        <bloque 2>
        
* **error** es un parámetro opcional, permite capturar sólo el tipo de error indicado, si no se coloca captura todos los errores posibles (es buena práctica capturar cada tipo de error por separado)
* **as alias** nos permite crear un alias al error, para trabajar con él.  

Entonces el código puede escribirse de la siguiente manera:

````python
def palindrome(string):
    return string == string[::-1]

try:
    print(palindrome(1))    

except TypeError as te:
    print("Solo se pueden ingresar strings") 
````

Tenemos la posibilidad de crear nuestros propios errores dentro de Python con la funcipon **raise**. Por ejemplo, en el programa anterior, si el usuario ingresa un string vacío, osea nada, Python no lo detectaría como un error, pero nosotros sabemos que "nada" no puede ser considerado como un palíndromo. Para evitar eso usamos la función **raise**

````python
def palindrome(string):
    try:
        if len(string) == 0:
            raise ValueError("No se puede ingresar una cadena vacía")
        return string == string[::-1]
    except ValueError as ve:
        print(ve)
        return False

try:
    print(palindrome(""))    
except TypeError as te:
    print("Solo se pueden ingresar strings")   
````

* **raise** Esta instrucción nos permite generar errores, es decir crear nuestros propios errores cuando detectemos que nuestro programa no actúa como debería con cierto tipo de datos

Su sintaxis es:

    raise <NombreError>("<descripci[on del error>")

Por último, existe la función **finally**. 

* **finally** Es una bloque de código que se ejecuta exista un error o no (un tercer bloque después de try except), no es muy usual pero puede darse para cerrar archivos, conexiones a BBDD o liberar recursos

# Assert Statements

Además de manejar errores con try, except y raise, podemos utilizar los **Assert Statements**, osea las afirmaciones en Python. El flujo de un **Assert Statement** es el siguiente:



Tenemos nuestro código, después llega el **Assert statement** y se ejecuta el código dentro de la afirmación, si es verdadero el código sigue, si es falso entonces devuelve un error.

La estructura es la siguiente:

    assert condición, mensaje de error
    
Básicmanete se lee como: "Afirmo que esta condición es verdadera, de lo contrario entrega este mensaje".

Ejemplo: 

Tenemos la siguiente función con código normal

````python
def palindrome(string):
    return string == string[::-1]

print(palindrome(""))
````

Podemos utilizar un **Assert Statement** para que devuelve un error en caso de que se ponga una cadena vacía.

Quedaría de la sigueinte manera

````python
def palindrome(string):
    assert len(string) > 0, "No se puede ingresar una cadena vacía"
    return string == string[::-1]
print(palindrome(""))
````

Como comprobamos, nos arroja el mensaje:

    AssertionError: No se puede ingresar una cadena vacía

Las afirmaciones son para errores del programador, no errores del usuario. Para los errores de los que se puede recuperar (como un archivo que no se encuentra o el usuario ingresa datos no válidos), genere una excepción en lugar de detectarla con una afirmación.

# ¿Cómo trabajar con archivos?

Dentro de la programación es muy común trabajar con distintos tipos de archivos. Existen dos categorías de archivos:

* **Archivos de text**: Tienen bytes que sólo representan letras, números o carácteres. (xml, txt, css, py, csv)
* **Archivos binarios**: Tienes bytes que representan datos más complejos como imágenes, audio, video, etc. (png, mp3, avi, dll)



Hagamos un programa para ver como funciona lo anterior.

Supongamos que tenemos un archivo .txt el cual contiene números en cada línea, y queremos almacenar todos estos números en una lista en python. 

````python
def read():
    numbers = []
    with open("./archivos/numbers.txt", "r", encoding="utf-8") as f:
        for line in f:
            numbers.append(int(line))
        print(numbers)
````

Con este código estamos abriendo el archivo *numbers.txt* en el modo de apertura de lectura "r" y lo nombramos como "f". La parte de *ecoding="utf-8"* nos ayudará a que Python no tenga problemas de interpretación con los carácteres latinos como las tildes. 

En este caso el archivo *numbers.txt* contiene lo siguiente:

    1
    45
    67
    872
    27
    5
    782
    214
    78
    456

Y al ejecutar el código se nos arroja una lista con todos esos números:

Ahora, digamos que tenemos una lista de nombres la cual queremos guardar en un archivo txt. Para ello utilizamos el siguiente código

    names = ["Erick", "Miguel", "Carlos", "Juan"]
    with open("./archivos/names.txt", "w", encoding="utf-8") as f:
        for name in names:
            f.write(name)
            f.write("\n")
            
Lo que este programa hará es crear un archivo llamado *names.txt* y agregará los nombres de la lista *names* en una línea nueva cada uno. Al correr el programa obtendremos los siguiente:

    Erick
    Miguel
    Carlos
    Juan
    
En caso de que se le hagan cambios a la lista y el programa se vuelva a correr, entonces el archivo *names.txt* se va a sobreescribir. Si no queremos que eso pase, utilizamos el modo de apertura "a", y todos los cambios se agregarán al final del archivo.   



En Python, normalmente sólo trabajarás con archivos de texto. 

Para abrir archivos dentro de Python existen tres modos de apertura:

* **R**: Lectura. Te permite leer todo lo que contenga el archivo.
* **W**: Escritura (sobreescribir). Elimina el archivo original y crea uno nuevo con la información agregada o editada.
* **A**: Escritura (agregar al final). No elimina el archivo original y agrega información al final de este.

Para escribir un archivo sin sobreescribir se hace con la siguiente sintaxis:

    with open("rutaDelArchivo", "r") as f:
* **with** es un manejador contextual. Se encarga de controlar el flujo del archivo en caso de que el programa se cierre o termine inesperadamente el archivo no se rompa
* **open** es la funcion de Python que permite abrir un archivo. Toma dos parámetros, el primero es la ruta del archivo y el segundo el modo de apertura (r, w o a)
* **as nombre** te permite identificar al archivo con un nombre que quieras. En este caso será el archivo "f".
